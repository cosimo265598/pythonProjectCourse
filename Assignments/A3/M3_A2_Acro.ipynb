# Acrobatic gymnastics

Acrobatic gymnastics is a competitive gymnastic discipline where teams of gymnasts work together and perform routines consisting of acrobatic moves, dance and tumbling, set to music.

We are interested in modelling all the aspects pertaining to an acrobatic gymnastic competition.

## Athletes, partnerships and categories

Each athlete is characterized by a first and last name, a gender, a date of birth and a nationality.
Athletes do not partake in the competition individually, but they form teams (also called partnerships or units).
There are five possible teams compositions:
- Women's pair (W2): a team composed of two female athletes 
- Women's group (W3): a team composed of three female athletes 
- Men's pair (M2): a team composed of two male athletes 
- Men's group (M4): a team composed of four male athletes 
- Mixed pair (MX): a team composed of a male and a female athletes

Each team competes in a single specific category, with specific age limits. Each athlete partakes in a single team.
For the purpose of this assignment, take into account only the following categories:
- Age-group 11-16 (11-16): all the team members must be at least 11 years old and at most 16
- Age-group 12-18 (12-18): all the team members must be at least 12 years old and at most 18
- Junior (J): all the team members must be at least 13 years old and at most 19
- Senior (S): all the team members must be at least 15 years old, without an upper bound

Rankings are split by category and team composition, e.g., a Junior MX team competes against only other Junior MX teams.
Each team in the competition is assigned a unique identifier.

Given the previous information, there could be up to 20 different rankings to be considered.

## Competition setup

The competition is organized in a qualifier round and a final round. All the teams partake in the qualifiers, but only a few are allowed to compete in the finals.
For the sake of simplicity, assume that only the first top-8 teams for each category/composition are allowed to move on.

Teams compete presenting their exercises (also called routines) to a judges panel. 

There are three types of exercises:
- Balance (B): an exercise with a focus on strength, poise and flexibility
- Dynamic (D): an exercise with a focus on throws, catches and tumbling elements
- Combined (C): a combination of balance and dynamic elements

Balance and dynamic exercises are performed in qualifiers whereas combined exercises are performed only during finals.

## Exercise scoring

The judges panel compile a scorecard for each exercise, consisting of the following fields:
- Difficulty (DS): a value between 0.00 and 3.00
- Penalty (PEN): a value between 0.00 and 10.00
- Execution (E1 through E4): four values between 5.0 and 10.0
- Artistry (A1 through A4): four values between 5.0 and 10.0

Both execution and artistry are evaluated taking into account only the average of two mid values, i.e., discarding the highest and lowest score for each of them, in order to compensate potential bias. The average scores determine the actual execution score (ES) and artistic score (AS). 

The overall score of the exercise is computed as: $TOT = DS - PEN + 2*ES + AS$

## Rankings

Rankings take into account only the overall score of each exercise, unless there are tie-breaks.
For qualifiers, the overall score of a given team is obtained taking into account the sum of the overall scores for its dynamic and balance exercise.
For finals, the overall score depends only on the combined exercise. The scores obtained during qualifiers are completely ignored for the final ranking.

### Tie-break rules

In case of ties, the following rules apply:
- If the overall score of two exercises is the same, the exercise with the highest execution score come first
- If there is still a tie, the exercise with the highest sum of execution and artistic scores (i.e., ignoring DS and PEN) come first
- If there is still a tie, the team with the smallest gap (in days) between youngest and oldest member come first
- If there is still a tie, it is resolved randomly through a coin flip
